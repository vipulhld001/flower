```python
import pandas as pd
import seaborn as sns
import numpy as np  # for test data only


rows = 100
data = {'Datasets': np.array(["CIFAR10", "MNIST", "CIFAR100", "FMNIST"], size=8),
        'school': np.random.choice(['Workstation', 'Collab', ], size=8)}
df = pd.DataFrame(data)
```

```python
p = sns.countplot(data=df, x='Datasets', hue='school')
p.legend(title='Machine Used', bbox_to_anchor=(1, 1), loc='upper left')

for c in p.containers:
    # set the bar label
    p.bar_label(c, fmt='%.000f', label_type='edge')
```

```python

import numpy as np  
import matplotlib.pyplot as plt  
from matplotlib.pyplot import figure
figure(figsize=(9, 12), dpi=80) 
X = ["CIFAR10", "MNIST", "CIFAR100", "FMNIST"] 
Ygirls = [635.65,562.51,641.14,557.56] #Workstation FedAvg 
Zboys = [1638.73,1576.74,1617.62,1504.77] #Collab FedAvg
Ygirlsx = [589.74,515.81,611.08,519.84] #   Workstation FedLR
Zboysx = [1528.82,1438.09,1490.00,1374.31] #Collab FedLR 
prox = [622.28,531.81,628.44,537.23] #Workstation FedProx
proy = [1662.25,1553.16,1637.13,1516.53] #Collab FedProx
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis + .14, Ygirls, 0.1, label = 'Workstation FedAvg') 
plt.bar(X_axis + .05, Zboys, 0.1, label = 'Collab FedAvg') 
plt.bar(X_axis - .05, Ygirlsx, 0.1, label = 'Workstation FedLR') 
plt.bar(X_axis - .14, Zboysx, 0.1, label = 'Collab FedLR') 
plt.bar(X_axis + .24, prox, 0.1, label = 'Workstation FedProx')
plt.bar(X_axis - .24, proy, 0.1, label = 'Collab FedProx')
plt.xticks(X_axis+.15, X) 
plt.xlabel("Datasets") 
plt.ylabel("Total Communication Time (sec)") 
plt.title("Total Communication Time for Datasets") 

plt.legend(fontsize='x-small',ncol=3) 


plt.show() 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
   
Women = [115, 215, 250, 200] 
Men = [114, 230, 510, 370] 
  
n=4
r = np.arange(n) 
width = 0.25
  
  
plt.bar(r, Women, color = 'b', 
        width = width, edgecolor = 'black', 
        label='Women') 
plt.bar(r + width, Men, color = 'g', 
        width = width, edgecolor = 'black', 
        label='Men') 

  
plt.xlabel("Year") 
plt.ylabel("Number of people voted") 
plt.title("Number of people voted in each year") 
  
# plt.grid(linestyle='--') 
plt.xticks(r + width/2,['2018','2019','2020','2021']) 
plt.legend() 
  
plt.show() 
```
