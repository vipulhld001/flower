```python
import pandas as pd
import seaborn as sns
import numpy as np  # for test data only


rows = 100
data = {'Datasets': np.array(["CIFAR10", "MNIST", "CIFAR100", "FMNIST"], size=8),
        'school': np.random.choice(['Workstation', 'Collab', ], size=8)}
df = pd.DataFrame(data)
```

```python
p = sns.countplot(data=df, x='Datasets', hue='school')
p.legend(title='Machine Used', bbox_to_anchor=(1, 1), loc='upper left')

for c in p.containers:
    # set the bar label
    p.bar_label(c, fmt='%.000f', label_type='edge')
```

```python

import numpy as np  
import matplotlib.pyplot as plt  
  
X = ["CIFAR10", "MNIST", "CIFAR100", "FMNIST"] 
Ygirls = [635.65,562.51,641.14,557.56] 
Zboys = [1638.73,1627.74,1710.00,1621.47] 
Ygirlsx = [589.74,515.81,611.08,519.84] 
Zboysx = [1528.82,1528.57,1587.59,1501.57] 
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.05, Ygirls, 0.09, label = 'Workstation FedAvg') 
plt.bar(X_axis + 0.1, Zboys, 0.09, label = 'Collab FedAvg') 
plt.bar(X_axis - 0.2, Ygirlsx, 0.09, label = 'Workstation FedLR') 
plt.bar(X_axis + 0.2, Zboysx, 0.09, label = 'Collab FedLR') 
plt.xticks(X_axis+.15, X) 
plt.xlabel("Datasets") 
plt.ylabel("Total Communication Time (sec)") 
plt.title("Total Communication Time for Datasets") 

plt.legend(fontsize='x-small',ncol=4) 
plt.show() 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
   
Women = [115, 215, 250, 200] 
Men = [114, 230, 510, 370] 
  
n=4
r = np.arange(n) 
width = 0.25
  
  
plt.bar(r, Women, color = 'b', 
        width = width, edgecolor = 'black', 
        label='Women') 
plt.bar(r + width, Men, color = 'g', 
        width = width, edgecolor = 'black', 
        label='Men') 

  
plt.xlabel("Year") 
plt.ylabel("Number of people voted") 
plt.title("Number of people voted in each year") 
  
# plt.grid(linestyle='--') 
plt.xticks(r + width/2,['2018','2019','2020','2021']) 
plt.legend() 
  
plt.show() 
```
